```python
import tensorflow as tf # machine learning
import cv2  # image processing
import numpy as np # matrix multiplication
import time # record query time
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
IMAGE_SIZE = (224, 224, 3)
model = tf.keras.models.load_model("/content/drive/MyDrive/HandClassification/HandClassification/Hand-Classification-For-Autism-Diagnosis/demo/MBNet")
```

```python
def predict(model, video):
    cap = cv2.VideoCapture(video)

    data_matrix = []

    # read in the frames
    i = 0
    while (cap.isOpened()):
        ret, frame = cap.read()
        if not ret: break

        image = cv2.resize(frame, (IMAGE_SIZE[0], IMAGE_SIZE[1])) # resize
        data_matrix.append(image)
        i += 1


    cur_time = time.time()
    prediction = model.predict(np.array([data_matrix]))
    # prediction = model.predict(data_matrix.clone().detach().to(device = 'cuda'))
    print(prediction)
    text_pred = "Hand Flapping" if prediction >= 0.5 else "No Hand Flapping"
    print(f"Prediction:  {text_pred} with {max([1 - prediction, prediction])} confidence.")

    cap.release()
    return time.time() - cur_time
```

```python
predict(model, "/content/drive/MyDrive/HandClassification/HandClassification/videos/val/ArmFlapping/video_3.mp4")
```
